```python
import easyocr
```

```python
#download the model
reader = easyocr.Reader(['en'], gpu = False)
```

```python
# display the image
import PIL
from PIL import ImageDraw
im = PIL.Image.open("imagesample.png")

im
```

```python
# Doing OCR. Get bounding boxes.
bounds = reader.readtext('imagesample.png')

bounds
```

```python
# Draw bounding boxes
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```
